```python
from scipy.stats import binom
import matplotlib.pyplot as plt
import numpy as np
```

```python
Coin = binom(100, 0.25)
x = np.linspace(0,100,101)
```

```python
plt.bar(x, Coin.pmf(x))
#print(Coin.pmf(x))
```

```python
import scipy.stats as stats
x2 = np.linspace(0,1,100)
a=2+1
b=2+4
rv = stats.beta(a,b)
plt.plot(x2, rv.pdf(x2)) #sample n=5
plt.plot(x2, stats.beta(2+12,2+38).pdf(x2)) #sample n=50
plt.plot(x2, stats.beta(2+25,2+75).pdf(x2)) #sample n=100
plt.plot(x2, stats.beta(2,2).pdf(x2))

print(np.argmax(rv.pdf(x2))/100)
```

```python
import scipy.stats as stats
x2 = np.linspace(0,1,100)
a=4+1
b=4+4
rv = stats.beta(a,b)
plt.plot(x2, rv.pdf(x2))
plt.plot(x2, stats.beta(2,2).pdf(x2))
print(np.argmax(rv.pdf(x2))/100)
```
