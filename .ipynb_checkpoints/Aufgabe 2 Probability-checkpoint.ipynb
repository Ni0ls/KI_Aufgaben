```python
from scipy.stats import binom
import matplotlib.pyplot as plt
import numpy as np
```

## Aufgabe 1

```python
y = np.linspace(0,6,7)
fy =  [0] * 7
pWürfel = 1/6

Auge1 = binom(1, 0.5)
Auge2 = binom(2, 0.5)
Auge3 = binom(3, 0.5)
Auge4 = binom(4, 0.5)
Auge5 = binom(5, 0.5)
Auge6 = binom(6, 0.5)
```

```python
#Aufgabe 1
for i in range(7):
    fy[i] = pWürfel*(Auge1.pmf(i)+Auge2.pmf(i)+Auge3.pmf(i)+Auge4.pmf(i)+Auge5.pmf(i)+Auge6.pmf(i))
```

```python
plt.bar(y, fy)
print(fy)
print(y)
```

# Aufgabe 3

```python
# array with 100 points equally spaced in [-1,1]
x = np.linspace(-8, 8, 100)
y = x*x
plt.plot(x, y);
```

```python
import scipy.stats as stats
rv = stats.norm(1, 2) # Mittelwert und Standardabweichung
# Wahrscheinlichkeit am Punkt x=0
print('Erwartungswert', rv.mean() )
print('Varianz', rv.var() )
print('Wahrscheinlichkeiten', rv.pdf(np.array([1,2,3])))
```

```python
#1.
rv = stats.norm(-1, 2)
#2
plt.hist(rv.rvs(size=1000), density=True, bins = 100)
plt.plot(x, rv.pdf(x))
```

# Aufgabe 4

3. Ja es sind am Rand Normalverteilungen

```python
rv2d = stats.multivariate_normal(mean=[0,0], cov=[[2, 1],[1, 2]])
x = rv2d.rvs(size=1000)
x1 = x[:,0]
x2 = x[:,1]
plt.hist(x1, density=True, bins=20);
plt.hist(x2, density=True, bins=20);
```

```python
plt.hist2d(x1,x2);
```
