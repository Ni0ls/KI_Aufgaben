```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d

x1 = x2 = np.linspace(-10, 10, 20)
X1, X2 = np.meshgrid(x1, x2)
f = np.array([(x1 - 1)**2 - 2*x1*x2 + 2*x2**2 + 1 for x1, x2 in zip(np.ravel(X1), np.ravel(X2))])
f = f.reshape(X1.shape)

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(x1, x2, f)
```

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

def fun(x1, x2):
    return (x1 - 1)**2 - 2*x1*x2 + 2*x2**2 + 1

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x1 = x2 = np.linspace(-10, 10, 20)
X1, X2 = np.meshgrid(x1, x2)
fs = np.array([fun(x1, x2) for x1, x2 in zip(np.ravel(X1), np.ravel(X2))])
F = fs.reshape(X1.shape)

ax.plot_surface(X1, X2, F)

ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('f(x1, x2)')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

N = 10
s = 0.5 # strength of noise
x = np.linspace(0, 1, N)
xx = np.linspace(0,1, 100)
e = np.random.normal(0, s, N)
y = np.sin(2*np.pi*x) + e
```

```python
plt.scatter(x,y)
#Regression
L2 = np.zeros(10)
for K in range(10):
    L = 0
    p = np.polyfit(x, y, K)
    y_reg = np.polyval(p, x)
 
    for i in range(9): #Berechnung des Quadradfehlers bei aktueller Reggression
        L = L + 0.5*(y_reg[i] - y[i])**2
                     
    L2[K] = L #Zwischenspeichern der Fehlerwerte bei unterschiedlichem K
    
    y_reg = np.polyval(p, xx)
    plt.plot(xx, np.sin(2*np.pi*xx))
    plt.plot(xx, y_reg)
```

```python
K = range(10)
L2 = np.sqrt(2*L2/10) #Normierung
plt.plot(K, L2) #Ellenbogen plot der Fehler ber Dimension des Reggressionspolynoms
```
